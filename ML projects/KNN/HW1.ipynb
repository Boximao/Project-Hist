### (iii) 

```python
# seperate into 0 and 1
df_0=df[df.Grade_of_S==0]
df_1=df[df.Grade_of_S==1]
```

```python
train_0=df_0.iloc[:70]
train_1=df_1.iloc[:140]
test_0=df_0.iloc[70:]
test_1=df_1.iloc[140:]
```

```python
# merge into training and testing sets
train_set=[train_0,train_1]
train=pd.concat(train_set)
test_set=[test_0,test_1]
test=pd.concat(test_set)
train.describe()
```

```python
train_x=train.iloc[:,:6]
train_y=train.iloc[:,6]
test_x=test.iloc[:,:6]
test_y=test.iloc[:,6]
test_x.describe()
```

```python
# testing error
result=[]
error=[]
for k in range(208,0,-1):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_x,train_y)
    y_pred=knn.predict(test_x)
    r=accuracy_score(test_y,y_pred)
    err=1-r
    result.append([k,r])
    error.append([k,err])
result
```

```python
df_error=pd.DataFrame(error)
df_error.columns=['k_value','error_rate']
plt.scatter(x=df_error.k_value,y=df_error.error_rate,alpha=0.8)
plt.xlim(max(df_error.k_value), 0)
plt.title('testing error')
```

```python
# training error
result_train=[]
error_train=[]
for k in range(208,0,-1):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_x,train_y)
    y_pred=knn.predict(train_x)
    r=accuracy_score(train_y,y_pred)
    err=1-r
    result_train.append([k,r])
    error_train.append([k,err])
error_train=pd.DataFrame(error_train)
error_train.columns=['k','error_rate']
plt.scatter(x=error_train.k,y=error_train.error_rate,alpha=0.8)
plt.xlim(max(error_train.k), 0)
plt.title('training error')
```

```python
# confusion matrix
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(train_x,train_y)
y_pred=knn.predict(test_x)
r=accuracy_score(test_y,y_pred)
con_mat=confusion_matrix(test_y,y_pred)
con_mat
```

```python
true_positive=con_mat[[1],[1]]/(con_mat[[1],[1]]+con_mat[[1],[0]])
print(f'True positive rate is {true_positive}')
true_negative=con_mat[[0],[0]]/(con_mat[[0],[0]]+con_mat[[0],[1]])
print(f'True negative rate is {true_negative}')
precision=con_mat[[1],[1]]/(con_mat[[1],[1]]+con_mat[[0],[1]])
print(f'Precision is {precision}')
F1=(true_positive*precision)/(true_positive+precision)
print(f'F1 score is {F1}')
```
