### Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
df.head()
```

```python
df['Amount']=scaler.fit_transform(df[['Amount']])
```

```python
df.head()
```

```python
df=df.drop(['Time'],axis=1)
df.head()
```

```python
df=df.drop_duplicates(keep='first')
```

```python
X=df.drop('Class',axis=1)
y=df['Class']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE(random_state=69)
X_res,y_res=sm.fit_resample(X,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.2,random_state=69)
```

## Apply ML algorithms for classification

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.metrics import plot_confusion_matrix
```

```python
LR=LogisticRegression()
```

```python
LR.fit(X_train,y_train)
y_pred=LR.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
plot_confusion_matrix(LR,X_test,y_test,normalize='all')
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda=LinearDiscriminantAnalysis()
```

```python
lda.fit(X_train,y_train)
```

```python
y_pred=lda.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
plot_confusion_matrix(lda,X_test,y_test,normalize='all')
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
GNB=GaussianNB()
GNB.fit(X_train,y_train)
y_pred=GNB.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
plot_confusion_matrix(GNB,X_test,y_test,normalize='all')
```

### Turns out that logistic regression gives the best prediction out of the 3 ml algorithms
